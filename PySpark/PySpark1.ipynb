```python
# Spark architecture:
This video provides a detailed explanation of the Spark Architecture (1:05-1:08), a crucial topic for anyone working with Spark, as it's a frequent discussion point in interviews (1:10-1:18).

Here's a summary and notes from the video:

Summary of Spark Architecture

# What is Driver?
The Spark architecture operates on a cluster (1:21-1:27) composed of multiple machines. These machines are divided into a Master node (2:43-2:45) and Worker nodes (2:47-2:51). The Master node hosts a Resource Manager (3:36-3:51), such as YARN, which is responsible for allocating resources. When a developer submits a Spark application, it first interacts with the Resource Manager (5:10-5:13), requesting a Driver (5:25-5:29). The Driver is a container (7:29-7:35) that runs the Spark application's main method, either a PySpark driver for Python applications or a JVM main method (8:57-12:02).

Once the Driver is established, it then requests Executors (12:31-13:22) from the Resource Manager. Executors are also containers (13:50-14:07) allocated on Worker nodes, each with specified memory and CPU cores, to perform parallel data processing (14:14-14:48). If the application uses User Defined Functions (UDFs) (15:57-16:05), Python Workers will be initiated within the Executor containers to handle Python-specific code (17:09-17:12). The Driver coordinates the tasks across these Executors (20:23-20:36). After the application completes, the Driver and all Executors shut down (20:43-20:46).

Key Notes on Spark Architecture:

Cluster Setup (1:34-2:33): A cluster consists of multiple machines connected via a network. The video gives an example of 10 machines, each with 20 CPU cores and 100GB RAM, totaling 200 cores and 1TB RAM for the cluster.
Master-Slave Architecture (2:38-3:10): The cluster operates on a master-slave model, with one Master node (the "owner") and multiple Worker nodes (the "servants").
Resource Manager (YARN) (3:36-4:25): Resides on the Master node and is responsible for managing and allocating resources for Spark applications.
Developer's Role (4:38-5:08): The developer submits a Spark application to the cluster.
Driver Container (7:29-7:52): The Resource Manager allocates a container on a Worker node (e.g., W5 in the video's example) for the Driver. This is the entry point for the Spark application (8:40-8:54).
Application Master vs. Driver (8:40-8:49, 11:13-11:16): The container running the Spark application is called the Application Master, and the JVM main method within it is referred to as the Application Driver.
PySpark vs. JVM Main Method (8:57-12:02): If writing code in Python, a PySpark driver is created, which internally uses a JVM main method through a Java wrapper. If writing directly in Scala or Java, only the JVM main method (Application Driver) is present.
Executors Allocation (12:31-14:51): After the Driver is set up, it requests additional containers, called Executors, from the Resource Manager to distribute the processing tasks. Each Executor receives a specified amount of memory and CPU cores.
User Defined Functions (UDFs) (15:57-17:51): Using UDFs in PySpark requires a Python Worker to be present within each Executor. This can lead to performance degradation as it introduces additional overhead. It's generally recommended to use built-in Spark functions over UDFs for better optimization.
Execution Flow (20:23-20:46): The Driver sends data to the Executors, which process it and return values. Once the work is done, the Driver and Executors are terminated.
ASCII Diagram of Spark Architecture

(Resource Negotiation)
      [ CLIENT / DRIVER NODE ] ----------------------------------------+
      |                                                                |
      |   User Code (main.py)                                          v
      |          |                                          [ CLUSTER MANAGER ]
      |          v                                          (Resource Manager)
      |   +-----------------------+                         -------------------
      |   |    PySpark Driver     |                         |  1. YARN        |
      |   |       (main)          |                         |  2. Mesos       |
      |   +----------|------------+                         |  3. Kubernetes  |
      |              | (Py4J)                               |  4. Standalone  |
      |              v                                      +---------|-------+
      |   +-----------------------+                                   |
      |   |      JVM Driver       |                                   | (Allocates Containers)
      |   | (SparkContext/Session)|                                   |
      |   +-----------------------+                                   |
      |              |                                                |
      |              | (Task Assignment)                              |
      +--------------|------------------------------------------------|--+
                     |                                                |
                     v                                                v
      +------------------------------------------------------------------+
      |                          WORKER NODES                            |
      |                                                                  |
      |   [ Worker Node 1 ]                [ Worker Node 2 ]             |
      |   (e.g., 20 Core, 100GB RAM)                                     |
      |                                                                  |
      |   +---------------------------+    +---------------------------+ |
      |   |      Executor (e1)        |    |      Executor (e2)        | |
      |   | (Container: 5 Core, 25GB) |    |                           | |
      |   |                           |    |                           | |
      |   |  +---------------------+  |    |  +---------------------+  | |
      |   |  |        JVM          |  |    |  |        JVM          |  | |
      |   |  | (Task/Cache/Shuffl) |  |    |  |                     |  | |
      |   |  +---------|-----------+  |    |  +---------------------+  | |
      |   |            | (Pipe)       |    |            |              | |
      |   |            v              |    |            v              | |
      |   |  +---------------------+  |    |  +---------------------+  | |
      |   |  |    Python Worker    |  |    |  |    Python Worker    |  | |
      |   |  |   (Executes Logic)  |  |    |  |                     |  | |
      |   |  +---------------------+  |    |  +---------------------+  | |
      |   +---------------------------+    +---------------------------+ |
      +------------------------------------------------------------------+
```

```python
# Parquet file format
Parquet works at column based file format not row based file format so it is faster. (see diagram you will understand more)
- Parquet use structured file format & Binary form and column based

Data In Columns On Disk (Logical View)
   +--------------------+--------------+-------+
   |       Title        |     Date     | Chart |
   +--------------------+--------------+-------+
   | Led Zeppelin IV    | 11/08/1971   |   1   |
   | Houses of the Holy | 03/28/1973   |   1   |
   | Physical Graffiti  | 02/24/1975   |   1   |
   +--------------------+--------------+-------+

   1. Row-Oriented data on disk
      (Stores data one complete row at a time)

      [ Led Zeppelin IV ] [ 11/08/1971 ] [ 1 ]  [ Houses of the Holy ] [ 03/28/1973 ] [ 1 ] ...
      ----------------------------------------  -------------------------------------------
                      Row 1                                      Row 2

   2. Column-Oriented data on disk
      (Stores all data for one column together)

      [ Led Zeppelin IV ] [ Houses of the Holy ] [ Physical Graffiti ] [ 11/08/1971 ] [ 03/28/1973 ] [ 02/24/1975 ] [ 1 ] [ 1 ] [ 1 ]                               

Goal: Write once & Read many.
Our goaal: Cost & Time reduce and performance high.

OLAP                    vs            OLTP
   (Online Analytical Processing)     (Online Transactional Processing)
   ------------------------------     ---------------------------------

   -> only few colums read            -> write
                                         ↳ insert, update, delete

   -> column based                    -> row based

# Row Based

       [ LOGICAL VIEW ]
       (Table Structure)

          Col A   Col B   Col C
         +-----+ +-----+ +-----+
   Row 0 |  A0 | |  B0 | |  C0 |  ✔ ✔ ✔ (Selected)
         +-----+ +-----+ +-----+
   Row 1 |  A1 | |  B1 | |  C1 |
         +-----+ +-----+ +-----+
   Row 2 |  A2 | |  B2 | |  C2 |
         +-----+ +-----+ +-----+
   Row 3 |  A3 | |  B3 | |  C3 |
         +-----+ +-----+ +-----+
   Row 4 |  A4 | |  B4 | |  C4 |
         +-----+ +-----+ +-----+
   Row 5 |  A5 | |  B5 | |  C5 |
         +-----+ +-----+ +-----+

                  |
                  | (Stored on Disk)
                  v

       [ PHYSICAL VIEW ]
   ✔    ✔    ✔
   +----+----+----+----+----+----+----+----+----+
   | A0 | B0 | C0 | A1 | B1 | C1 | A2 | B2 | C2 | ...
   +----+----+----+----+----+----+----+----+----+
   [    Row 0     ] [    Row 1     ] [    Row 2     ]

   +----+----+----+----+----+----+----+----+----+
   | A3 | B3 | C3 | A4 | B4 | C4 | A5 | B5 | C5 | ...
   +----+----+----+----+----+----+----+----+----+
   [    Row 3     ] [    Row 4     ] [    Row 5     ]

# Column Based

       [ LOGICAL VIEW ]
       (Table Structure)

          Col A   Col B   Col C
         +-----+ +-----+ +-----+
   Row 0 |  A0 | |  B0 | |  C0 |
         +-----+ +-----+ +-----+
   Row 1 |  A1 | |  B1 | |  C1 |
         +-----+ +-----+ +-----+
   Row 2 |  A2 | |  B2 | |  C2 |
         +-----+ +-----+ +-----+
   Row 3 |  A3 | |  B3 | |  C3 |
         +-----+ +-----+ +-----+
   Row 4 |  A4 | |  B4 | |  C4 |
         +-----+ +-----+ +-----+
   Row 5 |  A5 | |  B5 | |  C5 |
         +-----+ +-----+ +-----+

                  |
                  | (Transformation)
                  v

       [ PHYSICAL VIEW ]
       (Grouped by Column)

   (   Column A Block    )   (   Column B Block    )   (   Column C Block    )
   +--+--+--+--+--+--+---+   +--+--+--+--+--+--+---+   +--+--+--+--+--+--+---+
   |A0|A1|A2|A3|A4|A5|...|   |B0|B1|B2|B3|B4|B5|...|   |C0|C1|C2|C3|C4|C5|...|
   +--+--+--+--+--+--+---+   +--+--+--+--+--+--+---+   +--+--+--+--+--+--+---+

# Hybrid (e.g., Parquet) horizontal/vertical . 
# Suppose we have 1 mil record and we will take break when data is in 1 lakh chunk. Actually parquet use not column all not row all use hybrid model structure. 
       [ LOGICAL VIEW ]
       (Table Structure)

          Col A   Col B   Col C
         +-----+ +-----+ +-----+
   Row 0 |  A0 | |  B0 | |  C0 |  \
         +-----+ +-----+ +-----+   |
   Row 1 |  A1 | |  B1 | |  C1 |   |-> ROW GROUP 1 (Rows 0-2)
         +-----+ +-----+ +-----+   |
   Row 2 |  A2 | |  B2 | |  C2 |  /
         +-----+ +-----+ +-----+
   Row 3 |  A3 | |  B3 | |  C3 |  \
         +-----+ +-----+ +-----+   |
   Row 4 |  A4 | |  B4 | |  C4 |   |-> ROW GROUP 2 (Rows 3-5)
         +-----+ +-----+ +-----+   |
   Row 5 |  A5 | |  B5 | |  C5 |  /
         +-----+ +-----+ +-----+

                  |
                  | (Transformation)
                  v

       [ PHYSICAL VIEW ]
       (Row Groups on Disk)

   -------------------------------------------------------
   ROW GROUP 1 (Rows 0-2)
   [ A0  A1  A2 ]  [ B0  B1  B2 ]  [ C0  C1  C2 ]
   (Col A Chunk)   (Col B Chunk)   (Col C Chunk)
   -------------------------------------------------------
             ^
             | (Split on Disk)
             v
   -------------------------------------------------------
   ROW GROUP 2 (Rows 3-5)
   [ A3  A4  A5 ]  [ B3  B4  B5 ]  [ C3  C4  C5 ]
   (Col A Chunk)   (Col B Chunk)   (Col C Chunk)
   -------------------------------------------------------  

# Optimization Technique

       Query:
       SELECT * FROM table WHERE age < 18
                                 --------

       Example Data:
       Aadhar Card  (1.4 Billion Records)
       -----------

       Format:
       Parquet  (Stores Metadata)
       -------

       [ Row Group Level Metadata ]
       (Spark checks this BEFORE reading the actual data)


   Row Group 0   |   age min: 5    max: 25   |  -->  ✔ (KEEP)
                 |   (Could contain < 18)    |       (Read this block)


   Row Group 1   |   age min: 10   max: 17   |  -->  ✔ (KEEP)
                 |   (Definitely has < 18)   |       (Read this block)


   Row Group 2   |   age min: 22   max: 35   |  -->  ❌ (DISCARD)
                 |   (Lowest is 22. No match)|       (SKIP entirely!)   

# OPTIMIZATION TECHNIQUES
       -----------------------

   1. PREDICATE PUSHDOWN (Filter at Source) Skipping ROWS based on conditions

      Query:  SELECT * FROM users WHERE age < 20
                                        --------
                                       (Predicate)

      [ Parquet File on Disk ]
      ------------------------
      Row Group 1  |  Metadata: [min: 10, max: 15]  |  -->  ✔ READ
                   |  (10-15 is < 20? Yes.)         |       (Load to RAM)

      Row Group 2  |  Metadata: [min: 25, max: 40]  |  -->  ❌ SKIP
                   |  (25-40 is < 20? No.)          |       (Do not touch disk)

      Row Group 3  |  Metadata: [min: 18, max: 22]  |  -->  ✔ READ
                   |  (18-22 overlaps 20? Yes.)     |       (Load to RAM)


   2. PROJECTION PRUNING (Column Selection) Skipping COLUMNS that are not needed

      Query:  SELECT Name FROM users
                     ----
                  (Projection)

      [ Parquet File Layout ]
      -----------------------
      [  Column: Name  ]    [  Column: Age   ]    [  Column: City  ]
             ^                      ^                    ^
             |                      |                    |
        ✔ READ ONLY              ❌ SKIP              ❌ SKIP
      (IO = 100% used)         (IO = Saved)         (IO = Saved)                  
```

```python
# Data engineer interview question | Process 100 GB of data in Spark Spark | Number of Executors
for this case there is not a correct formula there (Dont understand lets see later)

1 cluster configuration was : 16 machine: 2 master(spark master, standly spark master: if one master failed we will start from 2nd master), 14 worker(w1,w2,w3,w4,w5,w6,w7,w8,w9,w10,w11,w12,w13,w14) & (128 RAM , 48 core cpu per machine lage hue the)
** Master node knows how much memory allocated to each worker. We will calculate only worker because whole workload goes to worker only not in master. 
                Total Available Resource:
                      14 worker * 128 RAM = 1792 GB = 1.8 TB RAM
                      14 worker * 48 core = 672 core
Lets assume there are total 5 team & each team some no of application running(App1 = one application). ex:
Here 18 Application = 18 task/project going from whole team within 1 cluster.    
Team 1               Team 2           Team 3           Team 4           Team 5     
+-------------+  +-------------+  +-------------+  +-------------+  +-------------+
|             |  |             |  |             |  |             |  |             |
| +---------+ |  | +---------+ |  | +---------+ |  | +---------+ |  | +---------+ |
| |  APP 1  | |  | |    1    | |  | |         | |  | |         | |  | |  APP 1  | |
| +---------+ |  | +---------+ |  | |         | |  | |         | |  | +---------+ |
|             |  |             |  | |         | |  | |         | |  |             |
|             |  | +---------+ |  | |         | |  | |         | |  | +---------+ |
|             |  | |    2    | |  | |  APP 1  | |  | |         | |  | |  APP 2  | |
|             |  | +---------+ |  | |         | |  | |         | |  | +---------+ |
|             |  |             |  | |         | |  | |         | |  |             |
|             |  | +---------+ |  | |         | |  | |         | |  |      o      |
|             |  | |    3    | |  | |         | |  | |         | |  |             |
|             |  | +---------+ |  | +---------+ |  | |         | |  |             |
|             |  |             |  |             |  | |         | |  |             |
| +---------+ |  | +---------+ |  |             |  | |  APP 1  | |  |             |
| |  APP 2  | |  | |    4    | |  |             |  | |         | |  |             |
| +---------+ |  | +---------+ |  |             |  | |         | |  |             |
|             |  |             |  | +---------+ |  | |         | |  | +---------+ |
|             |  | +---------+ |  | |         | |  | |         | |  | |  APP 3  | |
|             |  | |    5    | |  | |         | |  | |         | |  | +---------+ |
|             |  | +---------+ |  | |         | |  | |         | |  |             |
|             |  |             |  | |         | |  | |         | |  | +---------+ |
|             |  | +---------+ |  | |  APP 2  | |  | |         | |  | |  APP 4  | |
|             |  | |    6    | |  | |         | |  | |         | |  | +---------+ |
|             |  | +---------+ |  | |         | |  | |         | |  |             |
|             |  |             |  | |         | |  | |         | |  |  +-------+  |
| +---------+ |  | +---------+ |  | |         | |  | |         | |  |  |   5   |  |
| |  APP 3  | |  | |    7    | |  | +---------+ |  | +---------+ |  |  +-------+  |
| +---------+ |  | +---------+ |  |             |  |             |  |             |
|             |  |             |  |             |  |             |  |             |
+-------------+  +-------------+  +-------------+  +-------------+  +-------------+                 


* Understand below solution first (LinkedIn): bahiyat solution. Doesnot work always this formula

100 GB = 800 partitions.(8*100)    (128MB ka 1 chunk hota hai-> so 1GB 8 partition happen)
800 partitions = 800 cores ❌ (we need 800 cores but not at a time)
4 cores per executor → 200 executors
Memory per executor: 
128 MB (default partition) × 4 = 512 MB per core RAM
So we have 4 cores per CPU → 512 × 4 = 2 GB / executor
So 200 executors with 4 core CPU and 2 GB RAM
Total = 400 GB RAM

Reverse Calculation:
1 core and 1 GB RAM reserved for worker node
(48 - 1) × 14 = 658 ≈ 650 cores
(128 - 1) × 14 = 1778 ≈ 1.8 TB
5 cores per executor → 650 / 5 = 130 executors
Average memory per executor = 1.8 TB / 130 ≈ 14 GB

Overhead exclusion (i.e. 10% of total, 384 MB): 
= 14 GB - 1.4 GB
= 12.5 GB / executor
5 core CPU

Actual way to Answer the above question:----
① Requirement gathering
   → Type of data : JSON/XML/tabular/csv/parquet 100gb
   → Type of transformation: any wide dependency/ simply select and dump anywhere
   → Single vs Multiple sources
   → Any SLA: (How much time this job will be comppleted 1hr/2 hr)
② Cluster configuration
(111)

3)Your Approach / Mindset
   ↓
   calculate the partitions : total 110 gb data
Table 1   100 GB = 800 Partitions
Table 2   10 GB  =  80 Partitions

Approach 1
15 GB RAM / Executor
5 core CPU
4 executors

### LINKEDIN POST:https://www.linkedin.com/posts/priyabrata-padhi_how-to-estimate-cpu-executors-memory-share-7403636494948823041-SFQE/
🚀 How to Estimate CPU, Executors & Memory to Process 25 GB Data in Spark
When tuning a Spark job, one of the most important skills is estimating the required CPU cores, number of executors, and memory based on the input size.
 Here’s a practical example using 25 GB of data.
✅ 1. How many CPU cores are required?
Spark processes data in partitions.
 Default partition size ≈ 128 MB.

25 GB = 25 × 1024 MB = 25,600 MB
Number of partitions = 25,600 / 128 = 200
🔹 Each partition is processed by one CPU core, so:
👉 Total CPU cores required = 200
✅ 2. How many executors do we need?
Let’s assume each executor is configured with 4 CPU cores (a common best practice).

Total executors = Total cores / Cores per executor
                 = 200 / 4
                 = 50 executors
👉 Executors required = 50
✅ 3. How much memory is needed per executor?
Rule of thumb:
 Memory per executor ≈ 4 × partition size = 4 × 128 MB = 512 MB per core
If each executor has 4 cores:

Executor memory = 4 × 512 MB = 2048 MB = 2 GB
👉 Per executor memory = 2 GB
✅ 4. What is the total memory required?

Total memory = Number of executors × Memory per executor
              = 50 × 2 GB
              = 100 GB
👉 Total cluster memory required = 100 GB

💡 Notes / Best Practices
- Spark defaults to 128 MB block size (HDFS-style partitioning).
- Typically, use 2–5 CPU cores per executor for optimal parallelism.
- Executor memory should be ≥ 1.5× Spark reserved memory; small   executors avoid JVM overhead.
- More partitions = better parallelism, but too many can cause scheduling 
overhead.

```

```python
from pyspark.sql.functions import udtf

@udtf(returnType="word STRING")
class udtf_class():
    def eval(self,text:str): #Function name to be used as eval
        for word in text.split():
            yield (word,)

```

```python
# whenever we have to use constraints have to use lit
from pyspark.sql.functions import lit

udtf_class(lit("Hello World, this is a developer")).show()
```
