```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType
from pyspark.sql.functions import col

df = {
 "product": {
 "id": 1,
 "details": {
 "name": "Laptop",
 "specs": {
 "processor": "Intel i7",
 "memory": "16GB"
 }
 }
 }
}

# Define the schema explicitly
schema = StructType([
    StructField("product", StructType([
        StructField("id", IntegerType(), True),
        StructField("details", StructType([
            StructField("name", StringType(), True),
            StructField("specs", StructType([
                StructField("processor", StringType(), True),
                StructField("memory", StringType(), True)
            ]), True)
        ]), True)
    ]), True)
])

# Create DataFrame with explicit schema
df_spark = spark.createDataFrame([df], schema=schema)

# Select the nested column using dot notation
df_spark.select(col("product.details.specs.processor")).show()
```

```python
def two_sum(nums,target):
    lookup = {}
    for i,num in enumerate(nums):
        compliment = target - num
        if compliment in lookup:
            return [lookup[compliment],i]
        else:
            lookup[num] = i
print(two_sum([2,7,11,5],9))
```
