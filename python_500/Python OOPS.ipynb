```python
                                                # Multi Threading
# Multi Threading: Lets say you have a programme/function. Now you want to run that function multiple time, lets say 10 times. You can say I am pro I can run loops/for loops but the problem is that loops runs in sequence. That means that loop will complete your first iterations then it will go to second, third...till end. So whats wrong on this , nothing wrong. Suppose what if you want to process all the 10 iterations together parallelly. So want to do like this thats why Multi Threading comes into the picture. 
# It actually creats Multi Threads , so you can process your data simoltaniously. Normally you can run your programme parallely. You do not need to wait for the previous one to complete and start the second one. This is the power of Multi Threading. 
# Ex: Data Engineer Prospective: Lets say you just build a databricks notebook, where you writing your pyspark code. You want to perform the same thing for different different table. So if you are ruuning through loops first your first data will loaded and second and third like this... lets say you have 100 tables. You will be just waiting for all your 99 tables to be completed to just completed the last one. What if I say run all the applications together. So just imagine the speed. almost 100 times faster right? thats why Multi Threading comes into picture.

# For Multi Threading import some library. it is so fast that to check it like time.sleep(). time.sleep(random.randint(1,10)) will not give exact value gives only None, so put in a variable then pass through sleep.
import time
import random
from concurrent.futures import ThreadPoolExecutor
# def my_func(p_x):
#     wait = random.randint(1,10)
#     time.sleep(wait)
#     print(f"I am {p_x}. I took {wait} seconds")
# my_func(101)  

# Like this processing time, for 100 tables 1st one may take 10 sec, 2nd one 1 minute, 3rd one 40 sec according to volume of the data. So It is just like a environment.
# so without Multi Threrading how it looks: Each table will take time to process one by one. Ex tables: orders, products, customers, reviews, cancels.        Ex:  [Without Multithreading]

# table = ["orders","products","customers","reviews","cancels"]
# def my_func(p_x):
#     wait = random.randint(1,10)
#     time.sleep(wait)
#     print(f"I am {p_x}. I took {wait} seconds")

# for p_x in table:
#     my_func(p_x)
# OP: I am orders. I took 5 seconds
# I am products. I took 4 seconds
# I am customers. I took 6 seconds
# I am reviews. I took 7 seconds
# I am cancels. I took 10 seconds

# max_workers = len(table), map: map will maps with func and tables. Just look at the output all tables processed now parallely not waiting for one to complete, orders and cancels processed in same time. because it didnot wait to complete the previous last one. This is power of Thread Pool Executor.
#                                 Ex: [With Multithreading]
table = ["orders","products","customers","reviews","cancels"]
def my_func(i):
    wait = random.randint(1,10)
    time.sleep(wait)
    print(f"I am {i}. I took {wait} seconds")

with ThreadPoolExecutor(max_workers=len(table)) as executor:
    executor.map(my_func,table)
# OP: I am products. I took 3 seconds
# I am reviews. I took 4 seconds
# I am customers. I took 7 seconds
# I am orders. I took 10 seconds
# I am cancels. I took 10 seconds

# still if you dont beleive all processed parallely and give all output  at once  without waiting for the previous one for 3 sec.                                   
# table = ["orders","products","customers","reviews","cancels"]
# def my_func(i):
#     wait = 3
#     time.sleep(wait)
#     print(f"I am {i}. I took {wait} seconds")

# with ThreadPoolExecutor(max_workers=len(table)) as executor:
#     executor.map(my_func,table)

# Another way to run Multi Threading. Using loops . Dont think like it will works as sequence it will works parallely.                              [Ex: With Multithreading Loops]
# table = ["orders","products","customers","reviews","cancels"]
# def my_func(i):
#     wait = 3
#     time.sleep(wait)
#     print(f"I am {i}. I took {wait} seconds")

# with ThreadPoolExecutor(max_workers=len(table)) as executor:
#     for i in table:
#         future = executor.submit(my_func,i)
```

```python
                                                #Generator yield
# lets define a func for top 10 value. Normally a function return right
def topten():
    return 5
value = topten() 
print(value)      # OP:5                                            
# Now instead of return if you put yield. it is a special function which will works as a generator
def topten():
    yield 5
value = topten()
print(value.__next__())  # OP: <generator object topten at 0xffcb9af91900>

# Diffreence: check output. After return it will stop but for yield it is not.
# What is __next__()?
# = If we want to fetch something from iterator use the __next__() function

# Why someonw will use generator?
= Lets say you are fetching 1000 records from database, So when you are fetching 1000 records or may be you want to print them all or may be process something from those records. So when you want to process all thos 1000 records will be loaded into memory. And we want one value at a time, that case use generator instead of fetching the entire list. By using yield you can fetch one by one value.
```
