```python
%sql
-- In interview they will not ask syntax they will ask what is FK

--                          What is Foreign Key?
-- 1) In layman's terms, A foreign key is a key that points to a primary key in another table. 
-- 2) Foreign keys are used to enforce database integrity.
-- 3) The foreign key constraint prevents invalid data form being inserted into the foreign key column. 

--                        General formula is here
-- Alter table ForeignKeyTable add constraint ForeignKeyTable_ForiegnKeyColumn_FK 
-- FOREIGN KEY (ForiegnKeyColumn) references PrimaryKeyTable (PrimaryKeyColumn)
-- EX
-- Step 2: Add foreign key constraint to tblPerson
ALTER TABLE tblPerson ADD CONSTRAINT tblPerson_GenderId_FK FOREIGN KEY (GenderId) REFERENCES tblGender(ID);
```
