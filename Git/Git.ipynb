```python
VVI:                                                # Git Rebase
You should use Rebase with precaition. Lets say you are in Master Branch and have head which is latest commit. from here you created a feature branch where you did some commits. In Master branch also you did some commits. Now instead of using merge we can use "Rebase". What is Rebase? = "Rebase" simply move particular/feature branch after again new commit of head branch.
                                 ●
                                 |
   ●     ●                       |
   |     |                  ●----● 
   ●     |                  |
   |     ●                  ●
   ●     |     -------->    |
   |     |                  ●
   ●----'                   |
   |                        ●--
   |                        |
   |                        |
 Master                    Master
So, Earlier it was originated from main branch and Now as we shifted it's base it is called Rebase.    

Advantage of Rebase?
= Now if you observed . If you apply Merge command after Rebasing it, it will become 'Merge FF' (Merge Fast Forward). That means we do not have any new commit in particular master branch. All the commits are coming from feature branch is called Fast-Forward-Merge 'Merge FF'. This Fast-Forward-Merge really looks clean whenever you want to watch the history. That is why people use combination of 'Rebase' & 'FF' merge. Merge conflict is also same here just like normal merge conflict resolution.

# Practical Rebase:
# create branch: 
git switch -c rebase_feat
Now open data_ingestion.txt add "I am ingesting big data":
git add . & git commit -m "ingest big data"
git switch master
# open git graph
* [ rebase_feat ] ingest big data : "One commit Ahead"
|
O   [ main ] finally conflict resolved: "One commit Behind"
|\
| \
open data_ingest.txt 
add text "I am ingesting batch data"
git add . & git commit -m "ingest batch data"
 # Now I want rebase feature branch
git switch rebase_feat
OP: Successfully rebased and updated refs/heads/rebase_feat. 
  #Now if I want merge
git switch main  & git merge rebase_feat  
OP: it is fast Forward branch
```

```python
                                            # Cherry Pick
Lets say you have Master branch and have some commits and you created a different branch and there 2 commits name bug fix & development. And you say bug fix is very important and it should be marge with Main.
Cherry Pick: As name suggests, we can actually pick any commit (from other branch) that we want to Merge on top of your Master branch.
            NB: during cherry pick of any commit to merge with Main the commit id will be different after merge with Main.      
                            (Dev Branch)
             |                  |
    abc1234  O <----------.    [O] development
             |             \    |
             O -------------\- [O] bug fix <--- bd jbbsscbj
             |                      
             O                      
             |                      
             O                   [ Cherry Pick ]
             |
             O 
      (Main Branch)  

(base) ritwik@Ritwiks-MacBook-Air RITWIK PRADHAN GIT % git switch rebase_feat
Switched to branch 'rebase_feat'     and add some thing in readme.md file and add this text '' and (base) ritwik@Ritwiks-MacBook-Air RITWIK PRADHAN GIT %  git add .
(base) ritwik@Ritwiks-MacBook-Air RITWIK PRADHAN GIT % git commit -m 'big fix'
[rebase_feat 3cf7c9e] big fix
 1 file changed, 1 insertion(+) also add data_ingestion.txt file 'I am a serious developer'  amd git add . & git commit -m 'development' now 'git switch main' and instead of merge only take 'big fix' This is called cherry picking.  So lets find commit id? = ' git switch rebase_feat'  and (base) ritwik@Ritwiks-MacBook-Air RITWIK PRADHAN GIT % 'git log --oneline' . Now got the id . 
 Perform cherry Pick: switch main branch
 git cherry-pick id 
 # for multiple id cherry pick make sure you put correct order
 git cherry-pick id1, id2, id3 

 suppose you did some commits in qa and you want that commit only in production. So that hot fix can be done using that particular id using cherrypick .                                                      
```
