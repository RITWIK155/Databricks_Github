```python
%md%
1st:
BIG O: Mainly Used in Most Interview.
                    (Most of the time we deal TIME COMPLEXITY)
TIME COMPLEXITY: ex: for one problem solution have 2 code . 1st code run in 15 sec. 2nd code run in 4 min. This is called TIME COMPLEXITY.
SPACE CPMPLEXITY: Maybe 1st code run in 15 sec but take huge memory. But although 2nd code take 4 min and take only less memory during execution.
O(1)>O(log n)>O(n)>O(n^2)
                    WORST CASE (BIG O)
Best Case: OMEGA: Ω, 
Average Case: THETA: θ,
Worst Case: OMICRON/ O/BIG O: O
Ex: Iterate a list to find element 1 , 4, 7: [1,2,3,4,5,6,7]
                                              Ω     θ     O
Finding 1 Best case, Finding 4 Average Case, Finding 7 Worst Case
                    BIG O : O(n) {Proportional}
So any item passing n time will called O(n).
# passed n here "print_items(n)" and it runs n time "for i in range(n)" so O(n)
def print_items(n):
    for i in range(n):
        print(i)
print_items(10)    
            DROP CONSTANTS BIG O: O(2n)/O(10n)->O(n)
So this run n+n=2n time , so time complexity O(2n) by dropping constants it will O(n).
def print_items(n):
    for i in range(n):
        print(i)
    for j in range(n):
        print(j)
print_items(10)   
                    BIG O: O(n^2) [LESS EFFICIENT TIME COMPLEXITY] {Loop within a Loop}
So when a loop inside a for loop then it will O(n^2). 
ex:item will print n*n=n^2 time, here 10*10=100 time will print.               
def print_items(n):
    for i in range(n):
        for j in range(n):
            print(i,j)
print_items(10)  
                    BIG O: NON DOMINANTS: O(n^2+n)->O(n^2)
So when a loop inside a for loop then it will O(n^2). 
ex:item will print n*n=n^2 time for 1st 2 loop O(n^2), here 10*10=100 time will print. For 3rd for loop k will print n time, O(n). So total Time Complexity: O(n^2)+O(n)-> O(n^2+n).For millions of data n will drop will be O(n^2).           
def print_items(n):
    for i in range(n):
        for j in range(n):  ->O(n^2)
            print(i,j)
    for k in range(n):      ->O(n)
        print(k)
print_items(10)
                    BIG O: O(1) [MOST EFFICIENT & OPTIMAL]  {Constant}
So if n increases the no of operations will be constant & BIG O(1)
def print_items(n):
    return n+n+n  
                    BIG O: O(log n) [EFFICIENT after O(1)]  {Divide & Conquer}
So if you have to find a element in a list first cut half & if the item still not there then cut half again & again until the itm is found. This is called O(log n). For sorting algorithm use O(n log n)
Ex: Find 1 from 8 item this list [1,2,3,4,5,6,7,8]
    1st cut: [1,2,3,4] [5,6,7,8]
            Now remove [5,6,7,8] as item is not there
    2nd cut: [1,2] [3,4]
            Now remove [3,4] as item is not there
    3rd cut: [1] [2]
            Now remove [2] as item is not there
    Now item is found.                         
    So total 3 time cut is required. 2^3=8 or log₂ 8=3 
                    Interview: DIFFERENT TERMS of Input Time Complexity
So, instead having single parameter having 2 parameter, O(a+b)                    
def print_items(a,b):
    for i in range(a):  ->O(a)
        print(i)
    for j in range(b):  ->O(b)
        print(j)    
                    BIG O: Lists
item:     [11,3,23,7]                
index:     0  1 2 3
my_list.append(17)
    [11,3,23,7,17] ->O(1)
    0   1 2  3 4
my_list.pop()
    [11,3,23,7] ->O(1)
     0  1 2  3
my_list.pop(0)
    [3,23,7] ->
    0  1  2
my_list.insert(0,11)
    [11,3,23,7]  ->O(n) adding n item list
     0  1 2 3
my_list.insert(1,'Hi')
    [11,'Hi',3,23,7]  ->O(n), you can say O(1/2 n) as it is middle
     0   1   2 3  4
-------------------
2nd:
oops(my learning) : jiti notes

# Define a class named 'student'
class student():
    # The __init__ method is the constructor that initializes the object's attributes
    def __init__(self, name, age):
        self.name = name    # Instance attribute
        self.age = age
    #self allows you to access the object
    def printabout(self):
        print(f"I am a student & {self.name} & {self.age}")
# Create an instance (or object) s1 of the 'student' class        
# s1 = student()
s1 = student("john",20)
# Call the 'printabout' method
s1.printabout()

OP:I am a student & john & 20
1. Class: A class is a blueprint or template for creating objects.It is logical entity & does not create memory. It defines a set of attributes and methods(structure and behavior) that created objects will have. 
Ex: (jiti note) First a car's template is created . Based on that multiple car is created, here car is the class.
NB: When self is inside one function assume its a class. CLASS(VVI in Data structures)
    __init__ : constructor
ex1: class Cookie:
        def __init__(self,car):
            self.color = color
    cookie_one = Cookie('green')
    cookie_two = Cookie('blue')
ex2: 
class Cookie:
    def __init__(self,color):
        self.color = color
    def get_color(self):
        return self.color
    def set_color(self,color):
        self.color = color
cookie_one = Cookie('green')
cookie_two = Cookie('blue')
print("cookie one color is: ",cookie_one.get_color())
print("cookie two color is: ",cookie_two.get_color())

cookie_one.set_color('yellow')

print("Now cookie one color is: ",cookie_one.get_color())
print("Now cookie two color is: ",cookie_two.get_color())

ex2 op: 
cookie one color is:  green
cookie two color is:  blue
Now cookie one color is:  yellow
Now cookie two color is:  blue

2. Object: An object is instance of a class.It is an real world entity which contains an address & takes up some spaces in memory.
Ex: a specific car.

Constructor: A constructor is a special method in a class that is automatically called when an object is created. In Python, the constructor method is named __init__.
If no constructor or any __init__ function are not  there that is Default Constructor: 
default constructor use to instantiate the object of the class by python.
self: self is a mandatory argument in class.self allows you to access the instance's or object attributes from the class. function.'self' refers to the instance of the class.     
what is __init(parameter a, parameterb)?
=  
3. Encapsulation: Binding code and data in single unit to do a specific job without revealing unnessary data is known as encapsulation.  
Ex: A capsule that is wrapped with different medicine.
BankAccount class, the attribute balance is intended to be private, meaning it should not be directly accessed from outside the class.
class BankAccount:
    def __init__(self, balance):
        self.balance = balance  # Private attribute
        print(f"Account created with balance: {self.balance}")
    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited {amount}. New balance: {self.balance}")
    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance: {self.balance}")
        else:
            print("Insufficient funds")
    def get_balance(self):
        print(f"Current balance: {self.balance}")
        return self.balance
# Create an instance (or object) of BankAccount
account = BankAccount(1000)
account.deposit(500)
account.withdraw(200)
print(account.get_balance())

OP:
Account created with balance: 1000
Deposited 500. New balance: 1500
Withdrew 200. New balance: 1300
Current balance: 1300
1300
4. Abstraction: Hiding internal details & showing functionality is known as abstraction.
Ex: For a car , we need to know how to run it not the internal process, it can be achieved using abstraction.
from abc import ABC, abstractmethod
# The abc module stands for "Abstract Base Classes" and is used to define abstract base classes and abstract methods.
# Abstract base class
class Car(ABC):
    @abstractmethod
    def start_engine(self):
        pass
    @abstractmethod
    def stop_engine(self):
        pass
# Concrete class
class Tesla(Car):
    def start_engine(self):
        print("Tesla engine started.")
    def stop_engine(self):
        print("Tesla engine stopped.")
# Concrete class
class BMW(Car):
    def start_engine(self):
        print("BMW engine started.")
    def stop_engine(self):
        print("BMW engine stopped.")
# Create instances of Tesla and BMW
tesla_car = Tesla()
bmw_car = BMW()
# Use the abstracted methods
tesla_car.start_engine()
tesla_car.stop_engine()
bmw_car.start_engine()
bmw_car.stop_engine()

OP:
Tesla engine started.
Tesla engine stopped.
BMW engine started.
BMW engine stopped.
5. Inheritance: It is the mechanism by which an object (which is called child class) is created using the defination of parent object. It provides code reusability.
Ex: From car class we inherit fOUR wheeler and two wheeler.
# Define a base class named 'Car'
class Car:
    def __init__(self, make, model):
        self.make = make
        self.model = model
    def display_info(self):
        print(f"Car Make: {self.make}, Model: {self.model}")
# Define a derived class named 'FourWheeler' that inherits from 'Car'
class FourWheeler(Car):
    def __init__(self, make, model, seating_capacity):
        super().__init__(make, model)  # Call the constructor of the base class. super() function used to call a method from the parent class Car.Commonly used in the __init__ method to ensure that the parent class is properly initialized.
        self.seating_capacity = seating_capacity
    def display_info(self):
        super().display_info()  # Call the display_info method of the base class
        print(f"Seating Capacity: {self.seating_capacity}")
# Define another derived class named 'TwoWheeler' that inherits from 'Car'
class TwoWheeler(Car):
    def __init__(self, make, model, type_of_handlebar):
        super().__init__(make, model)  # Call the constructor of the base class
        self.type_of_handlebar = type_of_handlebar
    def display_info(self):
        super().display_info()  # Call the display_info method of the base class
        print(f"Type of Handlebar: {self.type_of_handlebar}")
# Create instances of FourWheeler and TwoWheeler
four_wheeler = FourWheeler("Toyota", "Camry", 5)
two_wheeler = TwoWheeler("Yamaha", "MT-07", "Sport")
# Use the display_info method
four_wheeler.display_info()
two_wheeler.display_info()

OP:
Car Make: Toyota, Model: Camry
Seating Capacity: 5
Car Make: Yamaha, Model: MT-07
Type of Handlebar: Sport
6. Polymorphism: It is the process by which some code, data, method or object behaves differently under different situations.
----------------------------------------------------------
                    POINTER
Another thing in POINTER is: GARBAGE COLLECTION     
to get the address id(p)               
p = 11 (So p addressing 11 value that is a ponter nothing else) 

ex:
As Dictionary is mutable so it will change the value for both dict1 & dict2 but the address will be same . Integer is immutable. So in dsa dictionary both value will be changed for both nodes
dict1 ={'value': 11}
dict2 = dict1
print("Before updating dict1 value: ",dict1)
print("Before updating dict2 value: ",dict2)
print("address/pointer of dict1: ",id(dict1))
print("address/pointer of dict: ",id(dict2))
dict2['value'] = 22
print("After updating dict1 value: ",dict1)
print("After updating dict2 value: ",dict2)
print("address/points of dict1: ",id(dict1))
print("address/pointer of dict: ",id(dict2))

OP:
Before updating dict1 value:  {'value': 11}
Before updating dict2 value:  {'value': 11}
address/pointer of dict1:  280385951351168
address/pointer of dict:  280385951351168
After updating dict1 value:  {'value': 22}
After updating dict2 value:  {'value': 22}
address/points of dict1:  280385951351168
address/pointer of dict:  280385951351168    

#Integer, after num2 value is same for num1 
num1 = 11
num2 = num1
print("Before updating num1 value: ",num1)
print("Before updating num2 value: ",num2)
print("address/pointer of num1: ",id(num1))
print("address/pointer of dict: ",id(num2))
num2 = 22
print("After updating num1 value: ",num1)
print("After updating num2 value: ",num2)
print("address/points of num1: ",id(num1))
print("address/pointer of dict: ",id(num2))

#OP
Before updating num1 value:  11
Before updating num2 value:  11
address/pointer of num1:  11646632
address/pointer of dict:  11646632
After updating num1 value:  11
After updating num2 value:  22
address/points of num1:  11646632
address/pointer of dict:  11646984
------------------------------------                
                    LINKED LIST
#context Linked List. Time complexity is: O(1) or O(n)              
class LinkedList:
    def __init__(self,value):
        create new Node
    def append(self,value):
        create new Node
        add Node to end
    def pop(self):
    def prepend(self,value):
        create new Node
        add Node to beginning
    def insert(self,index,value):
        create new Node
        insert Node
    def remove(self,index):

1) Linked List does not have indexes
2) Contigous memory. (All values will be next side by side/ all are next to each other in memory)  
3) Its all called node. (11, 3,23,7 each are node)
4) 11-> 3->  23->  7->None  
   |          |
  Head      Tail

                    LINKED LIST(BIG O)    
previous Linked List: 11 -> 3 -> 23 -> 7 Now ADD 4 at lst
Now Linked List: 11 -> 3 -> 23 -> 7 -> 4                    
TC: O(1) [that will not dependent on no of operations, Because it doesnot matter how many items we have in the list]

If REMOVE 4 from last what will be TC?
        11 -> 3 -> 23 -> 7 -> 4
Now:    11 -> 3 -> 23 -> 7 
TC: O(n) [as to remove 4 we have to iterate for each item]

ADD 4 at first/front what will be TC?
        11 -> 3 -> 23 -> 7 
Now:    4 -> 11 -> 3 -> 23 -> 7         
TC: O(1) [one time iteration at first so O(1), Because it doesnot matter how many items we have in the list]

REMOVE 4 at first what will be TC?
        4 -> 11 -> 3 -> 23 -> 7 
Now:    11 -> 3 -> 23 -> 7  
TC: O(1) [Because it doesnot matter how many items we have in the list]

ADD 4 at middle (b/w 23 & 7) what will be TC?
        11 -> 3 -> 23 -> 7 
Now:    11 -> 3 -> 23 -> 4 -> 7
TC: O(n) [Because we have to iterate each item in list until find 23 & 7]

REMOVE 4 at middle what will be TC?
        11 -> 3 -> 23 -> 4 -> 7 
Now:    11 -> 3 -> 23 -> 7
TC: O(n) [Because we have to iterate each item in list until find 4]

LOOKUP in the liked list , ex: 23 have to find out?
        11 -> 3 -> 23 -> 7
TC: O(n) [i)Because we have to iterate each item in list until find 23, ii)If you go by index in 2 index it have still you have to go 2th index by iterating so O(n)]

                Deep Concept(How Linked List work Under the hood)

   11 -> 3 ->  23 ->  7 -> 4 -> None
   |     |                 |
  Head  next              Tail  
What the NODE (4->None) is?
= Node is not only the value(4) it also a pointer(None) together it called node. So essentially its a dictionary.
{
    "value": 4,
    "next": None
}
Ex: 
11->3->23->7->4 
head = {
    "value": 11,
    "next": {
        "value": 3,
        "next": {
            "value": 23,
            "next": {
                "value": 7,
                "next": {
                    "value": 4,
                    "next": None
                }
            }
        }
    }    
}

OP: print(head['next']['next']['value']) # 23

                        Linked List CONSTRUCTOR
Defination: See above
see top context of Linked List: append, prepand, insert everytime new node creatring. So instead of creating 4 times node we will create a class node .
Ex:
# Node creation
class Node:
    def __init__(self,value): # node constructor __init__
        self.value = value
        self.next = None
# After node , assign head & tail
class LinkedList:
    def __init__(self,value): # linked list constructor __init__
        new_node = Node(value)
        self.head = new_node
        self.tail = new_node
        self.length = 1
# Create 4 node     
my_linked_list = LinkedList(4)
print(my_linked_list.head.value)                

OP: 4
                    LL: Print List
# 11->3->23->7->None , temp = self.head means temp = 11                    
def print_list(self):
    temp = self.head
    while temp is not None:
        print(temp.value)                    
        temp = temp.next

                    LL: Append
11->3->23->7->none
Add 4
11->3->23->7->4->none
condition: head should not be empty/none, so if head part itself is empty on that time we will add 4 as node in at head as well tail . else self.tail.next = new node (new node will create for 4) and self.tail=new_node & increase the length +1(optional) and return True             
# code snippt
def append(self,value):
    new_node = Node(value)
    if self.head is None:
        self.head = new_node
        self.tail = new_node
    else:
        self.tail.next = new_node
        self.tail = new_node
    self.length += 1
    return True
# full code of append 

class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class LinkedList:
    def __init__(self, value):
        new_node = Node(value)
        self.head = new_node
        self.tail = new_node
        self.length = 1

    def print_list(self):
        temp = self.head
        while temp is not None:
            print("temp.value:", temp.value)
            temp = temp.next

    def append(self, value):
        new_node = Node(value)
        if self.length == 0:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        self.length += 1

# Create list with 1, then append 2
my_linked_list = LinkedList(1)
my_linked_list.append(2)
my_linked_list.print_list()
#output
temp.value: 1
temp.value: 2
                            LinkedList: Pop
11->3->23->7->4
now remove 4
11->3->23->7
condition: Edge case 1: head should not be empty/none, so if head part itself is empty on that time we will return 
Edge case 2: we only have 1 node 
approach : pre, head, temp all in head, so if temp.next is not none 

# This can be done using 3 different scenerio code for
# Scenerio 1 , if head & tail is none
def pop(self):
    if self.length == 0:
        return None 
# Scenerio 2, 
temp 
 |
11 -> 3 -> 23 -> 7 -> 4
 |
Pre

temp = self.head
pre = self.head
# So code for this

while(temp.next): # any vaue there/ temp.next true
    pre = temp
    temp = temp.next # after iterating and pre is 7 and temp is 4 then    
self.tail = pre #tail 7
self.tail.next = None    

# so if we have 2 or more items in the list code will come # Full code
def pop(self):
    if self.length == 0:
        return none
    temp = self.head
    pre = self.head
    while(temp.next):
        pre = temp 
        temp = temp.next
    self.tail = pre
    self.tail.next = None
    self.length -= 1 
    # when last no more node left after removing length coming as 0
    if self.length == 0:
        self.head = None
        self.tail = None
    return temp.value

my_linked_list = LinkedList(1)
my_linked_list.append(2)

# (2) Items - Returns 2 Node
print(my_linked_list.pop())
# (1) Item - Returns 1 Node
print(my_linked_list.pop())
# (0) Item - Returns Node
print(my_linked_list.pop())

                            LL: Prepend
                            
add element at first
#How to create node, by taking parameter value in new node
#create node
def prepand (self,value):
    new_node = Node(value)                         
# if single node is there ex: 4 , then Head & Tail both 4
if self.length == 0:
    self.head = new_node
    self.tail = new_node
# else more nodes then next node will be head of next , ex:
        head
         |
4->None, 11->3->23->7->None
new_node.next = self.head
self.head = new_node

head         tail
 |             |
 4->11->3->23->7->None 

# Full code:
def prepand(self,value):
    new_node = Node(value)
    if self.length == 0:
        self.head = new_node
        self.tail = new_node
    else:
        new_node.next = self.head
        self.head = new_node
    self.length +=1
    return true

my_linked_list = LinkedList(4)
my_linked_list.append(7)
my_linked_list.prepand(3)
my_linked_list.print_list()   
                        LL: Pop First
ex-1: 11->3->23->7->None 
OP: 3->23->7, 11->None
head & temp both are 11
if we remove 11 then visual will be 
temp 11->None , head 3->23->7->None
self.head = temp.next
if after removing length become 0 then tail will be None

def pop_first(self):
    if self.length == 0:
        return None
# for more node , see first ex
    temp = self.head
    self.head = selfhead.next    
    temp.next = None
    self.length -= 1                        
    if self.length == 0:
        self.tail = None
    return temp
my_linked_list = LinkedList(2)
my_linked_list.append(1) 

# 2 items - Returns 2 Node
print(my_linked_list.pop_first)
# 1 items - Returns 1 Node
print(my_linked_list.pop_first)
# 0 items - Returns 0 Node
print(my_linked_list.pop_first)

                        LL: Get

Return the node at index, temp, head 11 & tail 7
length: total number of nodes 
11->3->23->7->None
0.  1.  2. 3 
def get(self,index):
    if index < 0 or index >= self.length:
        return None
    temp = self.head
    for _ in range(index):
        temp = temp.next
    return temp.val
my_linked_list = LinkedList(0)
my_linked_list.append(1) 
my_linked_list.append(2)
my_linked_list.append(3)
print(my_linked_list.get(2))
                                LL: Set
11->3->23->7->None
index = 2, value = 4
11->3->4->7->None
def set_value(self,index,value):
    temp = self.get(index)
    if temp:    # if temp is not none
        temp.value = value
        return True
    return False
my_linked_list = LinkedList(11)
my_linked_list.append(3)
my_linked_list.append(23)
my_linked_list.append(7)
my_linked_list.set_value(2,4)
my_linked_list.print_list()

                             LL: Insert
11->3->23->7->None
index = 2, value = 4
11->3->4->23->7->None
def insert(self,index,value):
    if index < 0 or index > self.length: #index out of range 
        return False
    if index == 0:
        return self.prepend(value)
    if index == self.length:
        return self.append(value)
    new_node = Node(value)
    temp = self.get(index - 1)
    new_node.next = temp.next
    temp.next = new_node
    self.length += 1
    return True
my_linked_list = LinkedList(11)
my_linked_list.append(3)
my_linked_list.append(23)
my_linked_list.append(7)
my_linked_list.insert(2,4)
my_linked_list.print_list()
                             LL: Remove
11->3->23->7->None
index = 2
11->3->7->None
def remove(self,index):
    if index < 0 or index >= self.length:
        return None
    if index == 0:
        return self.pop_first() # pop first item eleminated
    if index == self.length - 1:
        return self.pop() # pop last item eleminated
    prev = self.get(index - 1) # pop the element middle need prev 3 & temp 23
    temp = prev.next
    prev.next = temp.next # 11->3->7,  3 is prev and 23 is temp & 7 is temp.next 
    temp.next = None      # 23->None, now only 3 item there as temp is None, 11->3->7
    self.length -= 1
    return temp.value
my_linked_list = LinkedList(11)
my_linked_list.append(3)
my_linked_list.append(23)
my_linked_list.append(7)
my_linked_list.remove(2)
my_linked_list.print_list()
                                    LL: Reverse (Very common Interview Question)
11->3->23->7->None
None<-7<-23<-3<-11 , #tail(11), head(7), imagine animation:  first temp = head and then head = tail, reverse tail = temp, 3 line of code
def reverse(self):
    temp = self.head
    self.head = self.tail
    self.tail = temp
    # after = temp.next # temp 11, after = 3
    before = None # before = None
    for _ in range(self.length):
        after = temp.next # 3, 23, 7, None
        temp.next = before # None, 11, 3, 23
        before = temp      # 11, 3, 23
        temp = after.      # 3, 23, 7
my_linked_list = LinkedList(11)
my_linked_list.append(3)
my_linked_list.append(23)
my_linked_list.append(7)
my_linked_list.reverse()
my_linked_list.print_list()

# OR Full Code 
# Reverse Linked List
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
class LinkedList:
    def __init__(self, value):
        new_node = Node(value)
        self.head = new_node
        self.tail = new_node
        self.length = 1
    def print_list(self):
        temp = self.head
        while temp is not None:
            print(temp.value)
            temp = temp.next
    def append(self, value):
        new_node = Node(value)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        self.length += 1
    def reverse(self):
        temp = self.head
        self.head = self.tail
        self.tail = temp
        before = None
        for _ in range(self.length):
            after = temp.next
            temp.next = before
            before = temp
            temp = after

my_linked_list = LinkedList(11)
my_linked_list.append(3)
my_linked_list.append(23)
my_linked_list.append(7)
my_linked_list.reverse()
my_linked_list.print_list()
7->23->3->11
     
# OR Gemini  code                                     
def reverse(self):
    # Initial List: 11 -> 3 -> 23 -> 7 -> None    
    # --- SWAP HEAD AND TAIL ---
    temp = self.head        # temp is now 11
    self.head = self.tail   # self.head is now 7 (The end becomes the start)
    self.tail = temp        # self.tail is now 11 (The start becomes the end)

    # --- PREPARE VARIABLES ---
    after = temp.next       # after is 3 (Look ahead to the next node)
    before = None           # before is None (There is nothing before the head)
    
    # Now we enter the loop to reverse the arrows one by one...   
    # Current State: temp = 11, before = None 
    # Loop 1: Processing Node 11  Goal: Make 11 point backward to None.
    for _ in range(self.length):
        after = temp.next      # after = 3    (Save the connection to the rest of the list)
        temp.next = before     # 11 -> None   (Reverse the arrow! 11 now points to None)
        before = temp          # before = 11  (Move 'before' marker up to the node we just  finished)
        temp = after           # temp = 3     (Move 'temp' marker to the next node to process)
        
        ## Loop 2: Processing Node 3 Goal: Make 3 point backward to 11
        # # Current State: temp = 3, before = 11
        # after = temp.next      # after = 23   (Save the connection to the rest of the list)
        # temp.next = before     # 3 -> 11      (Reverse the arrow! 3 now points to 11)
        # before = temp          # before = 3   (Move 'before' up)
        # temp = after           # temp = 23    (Move 'temp' up)

        # # Loop 3: Processing Node 23 Goal: Make 23 point backward to 3.
        # # Current State: temp = 23, before = 3
        # after = temp.next      # after = 7    (Save connection)
        # temp.next = before     # 23 -> 3      (Reverse arrow! 23 now points to 3)
        # before = temp          # before = 23  (Move 'before' up)
        # temp = after           # temp = 7     (Move 'temp' up)

        # # Loop 4: Processing Node 7 Goal: Make 7 point backward to 23.
        # # Current State: temp = 7, before = 23
        # after = temp.next      # after = None (We reached the end!)
        # temp.next = before     # 7 -> 23      (Reverse arrow! 7 now points to 23)
        # before = temp          # before = 7   (Move 'before' up)
        # temp = after           # temp = None  (Move 'temp' into the void)
        
        # Final Result: 7 -> 23 -> 3 -> 11 -> None
Iteration,Node (temp),after (Saved),New Arrow (temp.next),before (Moves up)
1,11,3,11 -> None,11
2,3,23,3 -> 11,3
3,23,7,23 -> 3,23
4,7,None,7 -> 23,7      

# # OR Gemini code2 
    # --- GEMINI STYLE REVERSE METHOD ---
    def reverse(self):
        # --- SETUP PHASE ---
        # Current List: 11 -> 3 -> 23 -> 7 -> None
        
        # 1. Swap the Head and Tail markers
        temp = self.head        # temp starts at the beginning (11)
        self.head = self.tail   # New Head is 7
        self.tail = temp        # New Tail is 11
        
        # 2. Setup the 'before' variable
        # Since 11 will become the tail, it must point to None.
        before = None
        
        # 'after' is not defined yet; we will do it inside the loop.

        # --- LOOP PHASE ---
        # We iterate 4 times (for 11, 3, 23, and 7)
        for _ in range(self.length):
            
            # Step A: Save the next node so we don't lose the list!
            # Loop 1: after = 3
            # Loop 2: after = 23
            # Loop 3: after = 7
            # Loop 4: after = None
            after = temp.next
            
            # Step B: Flip the arrow backwards
            # Loop 1: 11 -> None
            # Loop 2: 3 -> 11
            # Loop 3: 23 -> 3
            # Loop 4: 7 -> 23
            temp.next = before
            
            # Step C: Move 'before' to the current node
            # Loop 1: before = 11
            # Loop 2: before = 3
            # Loop 3: before = 23
            # Loop 4: before = 7
            before = temp
            
            # Step D: Move 'temp' to the next node (stored in Step A)
            # Loop 1: temp = 3
            # Loop 2: temp = 23
            # Loop 3: temp = 7
            # Loop 4: temp = None (Loop ends)
            temp = after

# --- EXECUTION ---
my_linked_list = LinkedList(11)
my_linked_list.append(3)
my_linked_list.append(23)
my_linked_list.append(7)

print("Original List:")
my_linked_list.print_list()

my_linked_list.reverse()

print("\nReversed List:")
my_linked_list.print_list()
------------------------------------ ------------------------------------ ---------------------------------
                    Leet code Interview Question

                    Find Middle Node
# cant do with length                     
slow → moves one node at a time (slow = slow.next)
fast → moves two nodes at a time (fast = fast.next.next)
slow and fast are two pointers (references) to nodes in your linked list.

def find_middle_node(self):
    slow = self.head
    fast = self.head
    
    while (fast is not None) and (fast.next is not None):
        slow = slow.next
        fast = fast.next.next
        
    # where return runs in relation to the while block and when the loop stops.
    return slow
# You cant write head is not None and head.next is not None because head is a fixed reference to the first node of the list.    
#OR
while True:
    # Stop if fast can't move 2 steps
    if (fast is None) or (fast.next is None):
        break
    slow = slow.next
    fast = fast.next.next

return slow
# suppose many numbers are there 20 numbers are there how this logic will find middle or secnd middle 
1 -> 2 -> 3 -> ... -> 20 (arrange beneath, so slow will return 11)
Iteration	slow fast
    Start	1	1
    1	    2	3
    2	    3	5
    3	    4	7
    4	    5	9
    5	    6	11
    6	    7	13
    7	    8	15
    8	    9	17
    9	    10	19
    10	    11	21 (None) → STOP
                        Has Loop
sol:1
def has_loop(self):
    slow = self.head
    fast = self.head
    while fast is not None and fast.next is not None:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True
    return False 
# OR Sol:2                          
def has_loop(self):
    slow = self.head
    fast = self.head
    while (fast is not None) and (fast.next is not None):
        slow = slow.next
        fast = fast.next.next
        if slow is fast:   # <- this would return True even for NO loop
            return True
    return False                        
-------------------------------------
# annotation    
This is type hinting (also called type annotation). It tells what type of value the function will return.
def create_loop(self, pos: int) -> bool:    

pos: int → the parameter pos should be an integer
-> bool → the method returns a boolean (True or False)
It does not change the behavior of the function.It’s only for:
✔ readability, ✔ clarity, ✔ IDE hints (PyCharm, VS Code), ✔ static type checking (mypy, pylint)
-------------------------------------
```
