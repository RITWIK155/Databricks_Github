## **Data masking -Unity Catalog Fucntion** 

```python
%sql
CREATE TABLE ext_catalog.ext_schema.employee
(
    id INT,
    name STRING,
    salary INT
);

INSERT INTO ext_catalog.ext_schema.employee
VALUES (1, 'John', 10000), (2, 'Mary', 20000), (3, 'Mike', 30000);

SELECT * FROM ext_catalog.ext_schema.employee
```

```python
%sql
-- creating salary column on masking
CREATE FUNCTION ext_catalog.ext_schema.maskingk(salary STRING)
  RETURN CASE WHEN is_account_group_member('admins') THEN salary ELSE '***-**-****' END;
```

```python
%sql
ALTER TABLE ext_catalog.ext_schema.employee ALTER COLUMN salary SET MASK ext_catalog.ext_schema.maskingk;
```
