````python
                                            ## JOIN 
Absolutely Ritwik! Here's a **simple example** to help you understand how different types of joins (`left`, `right`, `inner`, `outer`) work in pandas.
REMEMBER: A join doesn't remove columns — it just decides which rows to keep based on the how parameter (left, right, inner, outer).
---

### 🧪 Let's say we have two small tables:

#### 🟦 Table A (`df_a`)
| ID | Name   |
|----|--------|
| 1  | Alice  |
| 2  | Bob    |
| 3  | Charlie|

#### 🟨 Table B (`df_b`)
| ID | City      |
|----|-----------|
| 2  | Mumbai    |
| 3  | Bangalore |
| 4  | Delhi     |

---

### 🔁 Now let’s see what each join does:

#### 1️⃣ **Left Join** (`how='left'`)
```python
pd.merge(df_a, df_b, on='ID', how='left')
```

✅ Keeps **all rows from Table A**  
🔗 Adds matching rows from Table B  
❌ If no match, fills with `NaN`

| ID | Name   | City      |
|----|--------|-----------|
| 1  | Alice  | NaN       |
| 2  | Bob    | Mumbai    |
| 3  | Charlie| Bangalore |

---

#### 2️⃣ **Right Join** (`how='right'`)
```python
pd.merge(df_a, df_b, on='ID', how='right')
```

✅ Keeps **all rows from Table B**  
🔗 Adds matching rows from Table A  
❌ If no match, fills with `NaN`

| ID | Name   | City      |
|----|--------|-----------|
| 2  | Bob    | Mumbai    |
| 3  | Charlie| Bangalore |
| 4  | NaN    | Delhi     |

---

#### 3️⃣ **Inner Join** (`how='inner'`)
```python
pd.merge(df_a, df_b, on='ID', how='inner')
```

✅ Keeps **only matching rows** from both tables

| ID | Name   | City      |
|----|--------|-----------|
| 2  | Bob    | Mumbai    |
| 3  | Charlie| Bangalore |

---

#### 4️⃣ **Outer Join** (`how='outer'`)
```python
pd.merge(df_a, df_b, on='ID', how='outer')
```

✅ Keeps **all rows** from both tables  
❌ Fills missing values with `NaN`

| ID | Name   | City      |
|----|--------|-----------|
| 1  | Alice  | NaN       |
| 2  | Bob    | Mumbai    |
| 3  | Charlie| Bangalore |
| 4  | NaN    | Delhi     |

---

Ah, got it Ritwik! You're referring to this example:

---


A **cross join** will combine **every row from Table A** with **every row from Table B**, regardless of the `ID`.

---

### ✅ Cross Join Result:

| ID_A | Name    | ID_B | City      |
|------|---------|------|-----------|
| 1    | Alice   | 2    | Mumbai    |
| 1    | Alice   | 3    | Bangalore |
| 1    | Alice   | 4    | Delhi     |
| 2    | Bob     | 2    | Mumbai    |
| 2    | Bob     | 3    | Bangalore |
| 2    | Bob     | 4    | Delhi     |
| 3    | Charlie | 2    | Mumbai    |
| 3    | Charlie | 3    | Bangalore |
| 3    | Charlie | 4    | Delhi     |

---

### 🧪 How to do this in pandas:

```python
import pandas as pd

# Define the tables
df_a = pd.DataFrame({
    'ID_A': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
})

df_b = pd.DataFrame({
    'ID_B': [2, 3, 4],
    'City': ['Mumbai', 'Bangalore', 'Delhi']
})

# Add dummy key to both
df_a['key'] = 1
df_b['key'] = 1

# Perform cross join
df_cross = pd.merge(df_a, df_b, on='key').drop('key', axis=1)

# Display result
print(df_cross)
```

---

Let me know if you'd like to visualize this or try it with your actual data!
````
